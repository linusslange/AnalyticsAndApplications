```python
import pandas as pd
```

```python
df_charging_sessions = pd.read_csv("./charging_sessions.csv")
print(df_charging_sessions.head())
```

```python
df_charging_sessions.info()
```

```python
# insgesamte Anzahl der Zeilen
anzahl_zeilen = df_charging_sessions.any(axis=1).sum()
print(f"Anzahl der Zeilen im Dataframe: {anzahl_zeilen}")

# Anzahl der Zeilen mit mindestens einem fehlenden Wert
anzahl_fehlende_werte = df_charging_sessions.isnull().any(axis=1).sum()
print(f"Anzahl der Zeilen mit fehlenden Werten: {anzahl_fehlende_werte}")
```

```python
df_null_werte = df_charging_sessions[df_charging_sessions.isnull().any(axis=1)]
df_null_werte.head(10)
```

```python
print(df_null_werte[df_null_werte["userID"].notnull() & df_null_werte["userInputs"].isnull()])
```

```python
spalten_liste = df_null_werte.columns.to_list()
print(spalten_liste)
for spalte in spalten_liste:
    anzahl_null_pro_spalte = df_null_werte[spalte].isnull().sum()
    print(f"In Spalte {spalte} gibt es {anzahl_null_pro_spalte} Null-Spalten")
```

```python
df_null_doneChargingTime = df_null_werte[df_null_werte["doneChargingTime"].isnull()]
df_null_doneChargingTime.head()
```
